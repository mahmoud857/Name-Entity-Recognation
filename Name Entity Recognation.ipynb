```python
import spacy 
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
st_1=nlp('Apple to build a Hong Kong factory for $6 million')
```

```python
for token in st_1:
    print(token.text, end=' | ')
```

```python
for ent in st_1.ents:
    print(ent.text)     #print the entity
    print(ent.label)    #print the entity label
    print(ent.label_)   #print the entity label in string
    print(str(spacy.explain(ent.label_))) #print the entity label in string with explanation
    print(ent.start) #start index of the entity
    print(ent.end)   #end index of the entity
    print(ent.start_char) #start character index of the entity
    print(ent.end_char)   #end character index of the entity
    print('----------------------------------------')
```

another way to give the entity 

```python
for ent in st_1.ents:
    print(ent.text,':',ent.label_,':',str(spacy.explain(ent.label_)))
```

```python
from spacy.tokens import Span
```

```python
st_3=nlp('CPRO to build a U.K. factory for $6 million')
ORG=st_3.vocab.strings['ORG']
new_ent=Span(st_3,0,1,label=ORG)
st_3.ents=list(st_3.ents)+[new_ent]
```

```python
for ent in st_3.ents:
    print(ent.text,end='|')
    print(ent.label,end='|')
    print(ent.label_)
```
